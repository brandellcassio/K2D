```python
import sys
import os
sys.path.append('../')
from evaluation import EvaluationHandler

evaluation = EvaluationHandler()
evaluation.load_query_file()
```

```python
ranks = []
for i in run_again:
    ranked_matches = evaluation.mapper.get_matches(evaluation.evaluation_items[i-1].segments)
    fixed_matches = evaluation._fix_matches(ranked_matches)
    evaluation.evaluation_items[i-1].ranked_matches = fixed_matches
    evaluation.evaluation_items[i-1].find_query_match_gt()
    ranks += [(evaluation.evaluation_items[i-1].ranked_matches, evaluation.evaluation_items[i-1].qm_position)]
```

```python
result_file =  '../../data/results/result.v04.final.v01.jsonl'
```

```python
fdata = open(result_file,'w')
```

```python
i = 0

with open(current_analysis, 'r') as f:
    for line in f:
        data = json.loads(line)
        write_data = False
        if data["query_id"] in to_validate_to_zero:
            data["query_candidate_item"] = 0
            write_data = True
            
        if data["query_id"] in run_again:
            idx  = run_again.index(data["query_id"])
            query_matches, pos = ranks[idx]
            result = '[{}]'.format(','.join([x.to_json() for x in query_matches]))
            data["query_candidate_item"] = 0
            data["query_matches"] = json.loads(result)
            write_data = True
        
        if write_data:
            fdata.write('{}\n'.format(json.dumps(data)))
        else:
            fdata.write(line)

fdata.close()     
```
