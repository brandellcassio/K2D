```python
b_colors = ['maroon', 'orangered', 'gold', 'blue']
nb_colors = ['maroon', 'orangered', 'gold', 'blue']
b_levels = ['b = 0.5 / GT kws + K2D', 'b = 0.3 / GT kws + K2D', 'b = 0.8 / GT kws + K2D']
nb_levels = [ 'b = 0.5 / NaLIR kws + K2D', 'b = 0.3 / NaLIR kws + K2D', 'b = 0.8 / NaLIR kws + K2D']
b_keys = ['b=0.5', 'b=0.3', 'b=0.8']
nb_keys = ['N+b=0.5', 'N+b=0.3', 'N+b=0.8']
def plot_chart(result, title, file_name, loc=0, gt=True, bbox=(0.4, 0.5), start_range=10, end_range=100):
    result['points']=range(1,6)
    # data to plot
    #print(headers, legends, n_groups)
    # create plot
    levels = b_levels
    keys = b_keys
    colors = b_colors
    if not gt:
        levels = nb_levels
        keys = nb_keys
        colors = nb_colors
    
    fig, ax = plt.subplots()
    ax.margins(x=0,y=0.4, tight=False)
    ax.set_ylim(start_range/100, end_range/100)
    ax.use_stick_edges = False
    ax.set_xticks([x for x in range(1,6)])
    ax.set_yticks([x * 0.01 for x in range(start_range,end_range+1, 5)], minor=False)
    with_legend = True
    #plt.figure(dpi=100)
    #print(nalir_level_legend[idx], nalir_level[level])
    plt.plot('points', 'baseline', data=result, marker='', color=colors[-1],\
    linewidth=2, label='Baseline/NaLIR', linestyle='dashed')
    
    for idx,key in enumerate(keys):
        if key == 'baseline':
            continue
        plt.plot('points', key, data=result, marker='', color=colors[idx],\
            linewidth=2, label=levels[idx])
    plt.xlabel('Ranking Position')
    plt.ylabel('% Precision')
    plt.title(title)
    #plt.xticks(index + (len(headers)/2)*bar_width, legends)
    if with_legend:
        plt.legend(loc='upper left', prop={"size":"13"},bbox_to_anchor=bbox, framealpha=1)
    plt.grid(True)
    plt.savefig(file_name)
    plt.tight_layout()
    plt.show()
```

```python
1#colors = ['maroon', 'orangered', 'gold', 'yellowgreen', 'blue']
b_colors = ['maroon', 'orangered', 'gold', 'yellowgreen',
          #, 'darkgreen','darkcyan', 'rebeccapurple', 'orchid', 
          'blue']

nb_colors = ['maroon', 'orangered', 'gold', 'yellowgreen', 'blue']

b_levels = ['b = 0.5 / GT kws + K2D',
            'b = 0.1 / GT kws + K2D', 
            'b = 0.02 / GT kws + K2D',
            'b = 0.008 / GT kws + K2D'
         ##   'b = 0.5 / NaLIR Keywords + K2D', 
         ##   'b = 0.1 / NaLIR Keywords + K2D', 
         ##   'b = 0.02 / NaLIR Keywords + K2D',
         ##   'b = 0.008 / NaLIR Keywords + K2D',
           ]

nb_levels = [
            'b = 0.5/NaLIR kws + K2D', 
            'b = 0.1/NaLIR kws + K2D', 
            'b = 0.02/NaLIR kws + K2D',
            'b = 0.008/NaLIR kws + K2D'
           ]

nb_keys = ['NaLIR+b=0.5', 'NaLIR+b=0.1', 'NaLIR+b=0.02', 'NaLIR+b=0.008', 'baseline']
b_keys = ['GT+b=0.5', 'GT+b=0.1', 'GT+b=0.02','GT+b=0.008','baseline']

def plot_chart(result, title, file_name, loc=0, gt=True, start_range=10, end_range=100):
    keys=b_keys 
    colors= b_colors
    levels = b_levels
    result['points']=range(1,6)
    
    if not gt:
        keys = nb_keys
        colors= nb_colors
        levels = nb_levels
    
    # data to plot
    #print(headers, legends, n_groups)
    # create plot
    levels += ['Baseline/NaLIR']
    fig, ax = plt.subplots()
    #ax.margins(x=0)
    #ax.set_xticks([x for x in range(1,6)])
    #ax.set_yticks([x * 0.01 for x in range(0,101, 3)])
    
    ax.margins(x=0,y=0.4, tight=False)
    ax.set_ylim(start_range/100, end_range/100)
    ax.use_stick_edges = False
    ax.set_xticks([x for x in range(1,6)])
    ax.set_yticks([x * 0.01 for x in range(start_range,end_range+1, 5)], minor=False)

    
    with_legend = True
    #plt.figure(dpi=100)
    #print(nalir_level_legend[idx], nalir_level[level])
    lines = []
    for idx,key in enumerate(keys):
        #print(key, result[key])
        if key == 'baseline':#or key.startswith('NaLIR+b'):
            #print(colors[idx%4])
            lines += plt.plot('points', key, data=result, marker='', color=colors[-1],\
    linewidth=2, label='Baseline/NaLIR', linestyle='dashed')
        else:
            #print(colors[idx])
            lines += plt.plot('points', key, data=result, marker='', color=colors[idx],\
                linewidth=2, label=levels[idx])
    plt.xlabel('Ranking Position')
    plt.ylabel('% Precision')
    plt.title(title)
    print(lines[4], levels[4:])
    #plt.xticks(index + (len(headers)/2)*bar_width, legends)
    if with_legend:
        #plt.legend(loc='lower right', prop={"size":"11"})
        ax.legend(lines[:2], levels[:2], loc='upper left', prop={"size": "13"}, framealpha=1)
        leg = Legend(ax, lines[2:], levels[2:], loc='upper left', prop={"size": "13"}, bbox_to_anchor=(0.29,0.4), framealpha=1)
        ax.add_artist(leg)
    plt.grid(True)
    plt.savefig(file_name)
    plt.tight_layout()
    plt.show()
```
